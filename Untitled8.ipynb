### Manipulating tensor (tensor operation)

Tensor operation include:
* Addition
* Subtraction
* Multiplication
* Division
* Matrix Multiplication


```python
tensor = torch.tensor([1,2,3])
tensor + 10
```

```python
#subtraction
tensor - 10
```

```python
tensor * 10
```

###Matrix Multiplication


```python
%%time
tensor.matmul(tensor)
```

```python
%%time
value = 0
for i in range(len(tensor)):
  value += tensor[i] * tensor[i]
print(value)
```

Two main rules for matrix multiplication

1. dimension cannot be same:
* (3,2),(3,2) will not work
* (2,3),(3,2) will work

2. the result matrix multiplication should give outer dimension:
*(2,3)@(3,2) -> (2,2)



```python
torch.matmul(torch.rand(3,2),torch.rand(2,3))
```
