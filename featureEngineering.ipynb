```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

pd.pandas.set_option('display.max_columns', None)
```

```python
dataset = pd.read_csv('train.csv')
dataset.head()
```

```python

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(dataset,dataset['SalePrice'],test_size=0.1,random_state=0)
```

**MISSING VALUE**

```python
#LET capture all the null value
#first lets handle Categorical feature which are missing

feature_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes=='O']

for feature in feature_nan:
  print("{}:{}% missing value".format(feature,np.round(dataset[feature].isnull().mean(),4)))
```

```python
#replace nan value with missing value

def replace_cat_feature(dataset,feature_nan):
  data=dataset.copy()
  data[feature_nan]=data[feature_nan].fillna("Missing")
  return data

dataset=replace_cat_feature(dataset,feature_nan)
dataset[feature_nan].isnull().sum()
```

```python
dataset.head()
```

```python
numerical_with_nan=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes!='O']

for feature in numerical_with_nan:
  print("{}:{}% missing value".format(feature,np.around(dataset[feature].isnull().mean(),4)))
```

```python
#Replace all the missing numerical missing vslue
for feature in numerical_with_nan:
  #replace by using median since their is oultiner
  median_value=dataset[feature].median()


  #create a new feature to capture null values
  dataset[feature+'nan']=np.where(dataset[feature].isnull(),1,0)
  dataset[feature].fillna(median_value,inplace=True)

dataset[numerical_with_nan].isnull().sum()
```

```python
dataset.head()
```

```python
#Temporal variables (Date Time Variable)

for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
  dataset[feature]=dataset['YrSold']-dataset[feature]
```

```python
dataset.head(10)
```

```python
dataset[["YearBuilt","YearRemodAdd","GarageYrBlt"]].head()
```

```python
dataset.head()
```

```python
import numpy as np
num_feature = ['LotFrontage','LotArea','1stFlrSF','GrLivArea','SalePrice']

for feature in num_feature:
  dataset[feature] = np.log(dataset[feature])
```

**Handling Rare Feature**

We remove categorical variab

```python
categorical_feature = [feature for feature in dataset.columns if dataset[feature].dtype == 'O']
```

```python
categorical_feature
```

```python
for feature in categorical_feature:
  temp = dataset.groupby(feature)['SalePrice'].count()/len(dataset)
  temp_df = temp[temp>0.01].index
  dataset.feature = np.where(dataset[feature].isin(temp_df),dataset[feature],'Rare_var')
```

```python
dataset.head(100)
```

**Feature Scaling**

```python
feature_scale = [feature for feature in dataset.columns if feature not in ['Id','SalePrice'] and dataset[feature].dtype!='O']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(dataset[feature_scale])
```

```python
scaler.transform(dataset[feature_scale])
```

```python
# transform the train and test set, and add on the Id and SalePrice variables
data = pd.concat([dataset[['Id', 'SalePrice']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(dataset[feature_scale]), columns=feature_scale)],
                    axis=1)
```
